```python
import numpy as np
import matplotlib.pyplot as plt

x1_0 = np.arange(1, 6)
x2_0 = np.array([1, 1, 3, 2, 4])

x1_1 = np.arange(6, 11)
x2_1 = np.array([9, 8, 9, 6, 7])

x1_for_line = np.arange(1, 11, 0.1)
w = -0.7
b = 10
y = w*x1_for_line + b


# w1 = ?
# w2 = ?
# b = ?

# w2*x2 = -w1*x1 - b


plt.scatter(x1_0, x2_0)
plt.scatter(x1_1, x2_1)
plt.scatter(x1_for_line, y)
plt.show()
```

서점에 보면 많은 책들이 있다. 페이지수(x1), 글자수(x2)로 장편소설과 단편소설을 구분한다면
