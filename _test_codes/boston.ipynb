```python
import tensorflow as tf
import matplotlib.pyplot as plt
(train_X, train_Y), (test_X, test_Y) = tf.keras.datasets.boston_housing.load_data()
tf.keras.datasets.reuters
print(train_X[:, 0:1].shape)
# print(len(train_X))
# print(train_X)

x_tags = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]

for x in range(train_X.shape[1]):
    plt.xlabel(x_tags[x])
    plt.ylabel("PRICE")
    plt.scatter(train_X[:, x], train_Y)
    plt.show()
```
